```python
### Input data
data=pd.DataFrame({'Dept':['Accounting','Accounting','Sales','Sales','Sales','Sales'],
                  'Employee':['Russel','Chris','Charlie','Ellen','Nina','Charlie'],
                  'Salary':[55000,66500,43500,60000,55000,43500]})
data
```

```python
 #Provide a brief description for the data. 
data.describe()
```

```python
#Sort the data by salary.
print(data)
data.sort_values(by=['Salary'])
```

```python
#Then sort the data by department and then by salary.
print(data)
data.sort_values(by=['Dept','Salary'])
```

```python
#Remove the duplicate record.
print(data)
data = data.drop_duplicates()
print(data)
```

```python
# Add a new column to indicate the income interest an employee needs to pay. The new variable Interest=Salary*0.05. 
print(data)
data = data.assign(Interest = data['Salary']*0.05)
data
```

```python
#d) Add a new column to indicate the salary level of each employee. 
#If an employee’s salary is higher than 60,000, then the salary level is high; 
#if the salary is higher than 50,000 but not higher than 60,000, then the salary level is medium; 
#if the salary is not higher than 50,000, then the level is low. 
## Step 1 - Define bins for salary

bins=[0,50000,60000,100000]
## Step 2- Name bins
bin_names=['Low','Medium','High']
## Divide salaries into bines
catg = pd.cut(data['Salary'], bins, labels=bin_names)
## Show summary of bins
levelcount=pd.value_counts(catg)
print(levelcount)
## Add as a new column
data = data.assign(SalaryLevel = catg)
print(data)
```

```python
#Show a summary of how many employees belong to each salary level category. 
levelcount=pd.value_counts(catg)
print(levelcount)
```

```python
#e) Create a new data frame by slicing the current data. 
#Only columns Dept, Employee, and Interest should remain in the new data frame. 
interest=data.loc[:,['Dept','Employee','Interest']]
interest
```

```python
#Then use the new data frame to calculate the mean of interests for each department.
grouped_dept = interest['Interest'].groupby(interest['Dept']).mean()
grouped_dept
```

```python
#■ Create a Data Frame
data={"Model": ["GA5", "GA6", "FLA4", "FLA9", "OL3", "GA6"],
"Price": [3000,2550,1880,5600,2700,2550],
"Sales": [75,23,98,66,27,23],
"Class": ['A','B','B','O','A','B']}
frame=pd.DataFrame(data, index=np.arange(1,7))
frame
```

```python
#Provide a brief description for the data.
frame.describe()
```

```python
#Sort the data by class and then by price. 
frame.sort_values(by=['Class','Price'])
```

```python
#Remove the duplicate record.
frame = frame.drop_duplicates()
frame
```

```python
#Add a new column to indicate the profit made from selling a model. The new variable Profit=PriceSales0.3.
frame = frame.assign(Profit = frame['Price'] * frame['Sales'] * 0.3)
frame
```

```python
#d) Add a new column to indicate the price level of each model. 
#If a model’s price is higher than 5,000, then the price level is high; 
#if the price is higher than 2,800 but not higher than 5,000, then the price level is medium; 
#if the price is not higher than 2,800, then the level is low. 
## Step 1 - Define bins for salary
bins=[0,2800,5000,10000]
## Step 2- Name bins
bin_names=['Low','Medium','High']
## Divide salaries into bines
catg = pd.cut(frame['Price'], bins, labels=bin_names)
## Show summary of bins
levelcount=pd.value_counts(catg)
print(levelcount)
## Add as a new column
frame = frame.assign(PriceLevel = catg)
print(frame)
```

```python
#e) Calculate the mean of profits for each price level.
grouped_price_level = frame['Profit'].groupby(frame['PriceLevel']).mean()
grouped_price_level
```
