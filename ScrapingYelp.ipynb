```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
quote_page=('https://www.yelp.com/search?cflt=restaurants&find_loc=Lubbock%2C+TX')
```

```python
page=urlopen(quote_page)
```

```python
soup=BeautifulSoup(page,'html.parser')
print(soup)
```

```python
all_star_box = soup.find_all('div', attrs={'class': 'i-stars__373c0__Y2F3O'})
star_list = []
for star_box in all_star_box:
    star_list.append(star_box.get('aria-label'))
```

```python
star_list
```

```python
df = pd.DataFrame(star_list, columns=["ratings"])
df
```
